**K-Means**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([7,3,5,7,9,2,4,1,9,5])
x2 = np.array([22,36,11,8,90,12,33,57,82,17])
```

```python
data = list(zip(x1,x2))
print(data)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2)
```

```python
model.fit(data)
plt.scatter(x1,x2,c = model.labels_)
plt.show() 
```

**iris.csv**

```python
import pandas as pd
df = pd.read_csv("iris.csv")
y1=df["petal.length"]
y2=df["petal.width"]
```

```python
data = list(zip(y1,y2))
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2)
model.fit(data)
plt.scatter(y1,y2,c = model.labels_)
plt.show() 
```

**Changing the number of clusters**

```python
data = list(zip(y1,y2))
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 4)
model.fit(data)
plt.scatter(y1,y2,c = model.labels_)
plt.show() 
```
